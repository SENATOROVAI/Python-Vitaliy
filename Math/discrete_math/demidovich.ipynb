2. Доказать что:

$$
a1 + а2 = а2 + а1; 
$$

 Чтобы это доказать мы вспомним что при сложении двух векторов складываюся их соотвестующие точки в определенном базисе а именно i,j,k. 

 Например у нас есть $$а1(5,3,2)$$ и   $$а2(4,6,8) $$ ,тогда $$а3 = (5+4,3+6,2+8),$$ $$a3(9,9,10)$$

 И при этом нет разницы в какой последовательности мы будем прибавлять значения в векторах , так как по свойствам сложения чисел 5+3 = 3+5 и так далее. 


3. Доказать что:
$$
a1+(a2+a3) = (a1+a2)+a3
$$

Мы можем это доказать следующим способом , зная о том что сложение векторов ничем собственно не отличается от обычного сложения чисел по своей сути, разница лишь в том что мы складываем не все числа 
подряд а только те которые имееют соответсвующие координатные точки, тогда согласно свойству 
сложения чисел если (5+3) + 2 = 10 , тогда  и 5 + (3+2) =  10.

Проверим в первом случае получается 8+2, а во втором 5+5 , обе эти операции дадут нам 10 в конце.


```python
"""Numpy это библиотека для работы с многомерными массивами."""

import numpy as np

# Создадим векторы а1 и а2
a1 = np.array([2, 3])  # Вектор a1: [2, 3]
a2 = np.array([5, 1])  # Вектор a2: [5, 1]

# Вычислим разность этих векторов
difference = a2 - a1

# Выводим векторы и их разность
print("Вектор a1:", a1)
print("Вектор a2:", a2)
print("Вектор a2 - a1:", difference)

# Проверка совпадает ли значения у переменной difference с результатом a2 - a1
if np.array_equal(difference, a2 - a1):
    print("Вектор a2 - a1 представляет собой разность векторов a2 и a1.")
else:
    print("Вектор a2 - a1 не представляет собой разность векторов a2 и a1.")
```

```python
# Создаем  вектор a
vector_a = np.array([2, 5, 7])

# Умножаем вектор vector_a на -1
negative_a = -1 * vector_a

# Проверяем, что negative_a равен вектору, противоположному a
result = np.array_equal(negative_a, -vector_a)

# Выводим векторы и результат
print("Вектор a:", vector_a)
print("Вектор -a:", negative_a)

if result:
    print("Умножение vector_a на -1 дает вектор, противоположный a.")
else:
    print("Умножение vector_a на -1 не дает вектор, противоположный a.")
```
