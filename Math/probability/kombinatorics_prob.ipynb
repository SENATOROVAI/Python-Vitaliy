```python
from itertools import accumulate, combinations, permutations
```

```python
fruit: list[str] = ["apple", "orange", "banana"]
list(permutations(fruit))
len([*permutations(fruit)])
```

```python
comb_int: int = len(
    [*combinations(fruit, 1)]), 
    len([*combinations(fruit, 2)]), 
    len([*combinations(fruit, 3)]
)
```

```python
comb_sum: int = sum(
    (
        len([*combinations(fruit, 1)]),
        len([*combinations(fruit, 2)]),
        len([*combinations(fruit, 3)]),
    )
)
```

```python
# Кумулятивная сумма
cummuative_sum: list[int] = list(accumulate([1, 2, 3]))
```

```python
def cumulative_sum(sequence: list[int]) -> list[int]:
    """
    Return the cumulative sum of a list of numbers.

    Args:
        sequence (list[int]): List of numbers.

    Returns:
        list[int]: Cumulative sum of the list of numbers.
    """
    cumulative_sum_inner: list[int] = []
    current_sum: int = 0

    for number in sequence:
        current_sum += number
        cumulative_sum_inner.append(current_sum)

    return cumulative_sum_inner


print(cumulative_sum([1, 2, 3]))
```
