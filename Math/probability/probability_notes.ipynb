# Выборочное пространство

- **Множество и обозначения:**
  - Множество — это совокупность объектов, которые называются точками.
  - Выборочное пространство (пространство точек): $\Omega$ (омега)
  - Точка выборочного пространства: $\omega$ (омега)

- **Примеры:**
  - $\omega_1, \omega_2, \omega', \ldots$ - различные точки
  - Подмножества: $S \subseteq \Omega$
  - Крайние случаи: $\Omega$ и пустое множество $\emptyset$

- **Размер множества:**
  - Обозначение: $|S|$
  - Значение: неотрицательное целое число или $\infty$
  - В частности, $|\emptyset| = 0$

- **Принадлежность точки множеству:**
  - $\omega \in S$ - точка принадлежит множеству $S$
  - $\omega \notin S$ - точка не принадлежит множеству $S$


Множество задается правилом членства точек выборочного пространства. Например, множества из (a′)–(f′) определены настолько корректно, насколько однозначно интерпретируются их словесные описания.

Один из способов задания множества - перечисление всех точек выборочного пространства, входящих в него. Однако такой способ может быть крайне трудоемким. Например, размер множества в примере (e) равен 66 = 46 656. 

Каждый исход бросания шести игральных костей можно представить в форме упорядоченной шестерки символов: $(s_1, s_2, s_3, s_4, s_5, s_6)$, где $s_j$, $j = 1, ..., 6$, принимают значения от 1 до 6. 


# Подмножество и надмножество

- **Подмножество:**
  - Обозначение: $S \subseteq \Omega$
  - Значение: множество точек выборочного пространства, входящих в множество $S$

- **Надмножество:**
  - Обозначение: $S \supseteq \Omega$
  - Значение: множество точек выборочного пространства, находящихся в множестве $S$

- **Примеры:**
- $\emptyset \subseteq \Omega$ - Пустое множество является подмножеством всего множества.
- $\Omega \supseteq \Omega$ - Множество содержит само себя (надмножество самого себя).
- $\Omega \subseteq \Omega$ - Множество является подмножеством самого себя.
- $S \subseteq \Omega$ - Множество $S$ является подмножеством множества $\Omega$.
- $S \supseteq \Omega$ - Множество $S$ содержит множество $\Omega$ (надмножество).
- $S \subsetneq \Omega$ - Множество $S$ строгое подмножество $\Omega$.
- $S \supsetneq \Omega$ - Множество $S$ строго больше множества $\Omega$ (надмножество).

* Строгое подмножество (обозначается как $A \subsetneq B$) означает, что множество $A$ является подмножеством множества $B$, но при этом не совпадает с $B$. Другими словами, строгое подмножество исключает возможность равенства множеств $A$ и $B$.

# Операции над множествами

- **Дополнение:**
    - Обозначение: $A^c$
    - Значение: множество точек выборочного пространства, не принадлежащих множеству $A$
    - Символически это записывается так: $A^c = { x \mid x \notin A }$
  
- **Обьединение:**
    - Обозначение: $A \cup B$
    - Значение: множество точек выборочного пространства, принадлежащих множеству $A$ или множеству $B$
    - Символически это записывается так: $A \cup B = { x \mid x \in A \text{ или } x \in B }$

- **Пересечение:**
    - Обозначение: $A \cap B$
    - Значение: множество точек выборочного пространства, принадлежащих обоим множествам $A$ и $B$
    - Символически это записывается так: $A \cap B = { x \mid x \in A \text{ и } x \in B }$

- **Разность:**
    - Обозначение: $A \setminus B$
    - Значение: множество точек выборочного пространства, принадлежащих множеству $A$, но не принадлежащих множеству $B$
    - Символически это записывается так: $A \setminus B = A \cap B^c = \{ x \mid x \in A \text{ и } x \notin B \}$
  
- **Симметрическая разность:**
    - Обозначение: $A \triangle B$
    - Значение: множество точек выборочного пространства, принадлежащих хотя бы одному из множеств $A$ и $B$
    - Символически это записывается так: $A \triangle B = (A \cap B^c) \cup (A^c \cap B) = (A \setminus B) \cup (B \setminus A)$


# Коммутативный закон

- **Объединение:**
    - $A \cup B = B \cup A$
    - Это значит, что объединение множеств $A$ и $B$ не зависит от порядка их объединения. То есть, множество всех элементов, принадлежащих хотя бы одному из множеств, будет одинаковым, независимо от того, с какого множества начинается объединение.

- **Пересечение:**
    - $A \cap B = B \cap A$
    - Это значит, что пересечение множеств $A$ и $B$ не зависит от порядка их пересечения. То есть, множество всех элементов, которые принадлежат обоим множествам, будет одинаковым, независимо от порядка множеств в операции пересечения.

# Ассоциативный закон

- **Объединение:**
    - $A \cup (B \cup C) = (A \cup B) \cup C$
    - Это значит, что способ группировки множеств при объединении не влияет на результат. Объединение трёх множеств всегда даст одно и то же множество всех элементов, принадлежащих хотя бы одному из этих множеств, независимо от порядка объединения.

- **Пересечение:**
    - $A \cap (B \cap C) = (A \cap B) \cap C$
    - Это значит, что способ группировки множеств при пересечении не влияет на результат. Пересечение трёх множеств всегда даст одно и то же множество всех элементов, принадлежащих всем трём множествам одновременно, независимо от порядка пересечения.

# Дистрибутивный закон

Операции объединения и пересечения связывает также пара дистрибутивных законов:

1. **Дистрибутивность пересечения относительно объединения:**
    - $A \cap (B \cup C) = (A \cap B) \cup (A \cap C)$ (D1)
    - Это значит, что пересечение множества $A$ с объединением множеств $B$ и $C$ эквивалентно объединению пересечений множества $A$ с множествами $B$ и $C$.
    - 
2. **Дистрибутивность объединения относительно пересечения:**
    - $A \cup (B \cap C) = (A \cup B) \cap (A \cup C)$ (D2)
    - Это значит, что объединение множества $A$ с пересечением множеств $B$ и $C$ эквивалентно пересечению объединений множества $A$ с множествами $B$ и $C$.


Сделаем несколько замечаний. Во-первых, аналогия с правилами арифметики справедлива для закона (D1):

- $A * (B + C) = (A * B) + (A * C)$ (D1)

но не верна в случае (D2):

- $A + (B * C) \neq (A + B) * (A + C)$ (D2) 

Следовательно, операция пересечения множеств обладает дистрибутивностью относительно операции объединения, что не всегда выполняется в арифметических выражениях.

# Законы Де Моргана

Три операции, определенные к этому времени: дополнение, объединение и пересечение, задействованы в двух следующих формулах, называемых законами Де Моргана:

1. **Закон Де Моргана для дополнения объединения:**
    - $(A \cup B)^c = A^c \cap B^c$
    - Это значит, что дополнение объединения множеств $A$ и $B$ эквивалентно пересечению дополнений множеств $A$ и $B$.

2. **Закон Де Моргана для дополнения пересечения:**
    - $(A \cap B)^c = A^c \cup B^c$
    - Это значит, что дополнение пересечения множеств $A$ и $B$ эквивалентно объединению дополнений множеств $A$ и $B$.

<br>

Из законов Де Моргана следует, что с помощью операции дополнения объединение можно выразить через пересечение и наоборот. В самом деле:

- $A \cap B = (A^c \cup B^c)^c$
- $A \cup B = (A^c \cap B^c)^c$

Нужно чтобы элементы из  $A$ не пересекались с элементами B, то есть разность $A /setminus B$ должна равняться $A$, а $B /setminus A$ равняться $B$. Если при разности мы получаем, что результат идентичен тому множеству относительно которого происходит сравнение, получаем, что ни одного элемента из A нет в B, из чего следует что данные множества не пересекаются.
