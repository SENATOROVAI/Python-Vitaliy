# <center> Базовые запросы </center>
SQL - простыми словами инструмент(язык) запросов для обращенияк базе  данных 

Базы данных -  это структура которая представляет собой хранилище наших данных , с помощью SQL мы делаем запросы к БД. 

Следующий запрос представляет дает задачу БД предоставить нам выборку всех данных с таблицы courier actions (select значит выбрать) (*  значит все данные) , (limit 10  значит ограничить выборку первыми 10ю записями в БД)
*cиним цветом в данном случае подсвечиваются команды  sql
Примечание: запросы лучше писать более явно, то есть вместо * нужно отдавать предпочтение названиям колонок в таблице бд.

![SELECT QUERY EXPL1](../Images/learn_sql_img/select_query_example1.png)


В продакшене не рекомендуется использовать запросы вроде "SELECT * FROM table;", так как большинство таблиц содержат миллионы строк данных. Этот тип запроса может привести к долгому выполнению или даже не выполниться вовсе. Поэтому важно явно указывать ограничение (LIMIT), если требуется получить результат запроса.

 
Следующий запрос отличается от предыдущего добавлением в него сортировки с помощью метода ORDER_BY + название таблицы по какой делаем сортировку , параметр DESC значит сделать выборку по убыванию, если его не указать, то выборка будет по возрастанию.

Всегда стоит явно указывать округление , даже если среда сразу показывает нужное количество цифр после запятой!

В оператор round первым параметром приходит число которое нужно округлить, а вторым значение до скольки знаков после запятой это следует сделать(если вторым параметром пришло положительное число), если же число наоборот отрицательное , то у нас  происходит округление перед запятой

![ROUND EXAMPLE](../Images/learn_sql_img/round_func_expl.png)
