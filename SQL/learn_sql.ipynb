<center> Группировка данных в SQL </center> 
<br>
Чтобы группировка сработала она должна выглядеть следующим образом
SELECT column_name  -- сначала выбираем колонку таблицы.
FROM table_name  -- затем указываем таблицу.
WHERE column_name = 1  указываем условие для фильтрации.
GROUP BY column_name  -- затем группируем по колонке.

То есть группировка всегда должна происходить после фильтрации данных.

Ключевое слово  HAVING используется когда нам нужно применить какое то условие к результатм агрегации (используется после WHERE)

Например вот как может выглядеть такой запрос:

SELECT product, SUM(amount) AS total_sales
FROM sales
WHERE amount > 50
GROUP BY product
HAVING SUM(amount) > 200;



Но что если в результат группировки и агрегации мы хотим включить не все группы, а только те, для которых выполняется некоторое условие? Например, только те размеры заказов, количество которых выше или ниже определённого значения? Или только те возрастные группы, количество пользователей в которых превышает заданный порог?

Судя по всему, для этого нам необходимо каким-то образом наложить фильтрацию уже поверх группировки. Но, как мы выяснили, оператор WHERE использовать после GROUP BY нельзя. Что же тогда делать?

Специально для таких случаев существует оператор HAVING, который тоже выполняет функцию фильтрации, но работает поверх уже сгруппированных данных.
<br>
<br>
<span style="color: #056744;">
В этом и заключается ключевое отличие операторов WHERE и HAVING: первый фильтрует записи в таблице перед группировкой и агрегацией, а второй — после. Кстати, это простой, но довольно популярный вопрос на собеседовании по SQL.
</span>
<br>
<br>

Ещё одна важная особенность работы оператора HAVING: в этом блоке нельзя указывать алиасы полей с результатом агрегации из блока SELECT. Дело в том, что в соответствии с порядком выполнения запроса оператор SELECT выполняется после оператора HAVING. Поэтому агрегацию необходимо также указывать и в HAVING.
<br>
<br>
<br>
<br>

Здесь важно понять, что группировка, агрегация и фильтрация групп со всеми расчётами происходят всегда до выполнения инструкций в блоке SELECT, где вы просто определяете, что попадёт в результирующую таблицу.

Таким образом, мы снова можем дополнить список ключевых слов в запросе:

```sql
SELECT
FROM
WHERE
GROUP BY
HAVING
ORDER BY
LIMIT
```

Порядок их выполнения в запросе следующий:

Сначала выполняется оператор FROM — происходит выбор нужной таблицы.
Далее WHERE — отфильтровываются строки, соответствующие условию.
Потом GROUP BY — строки объединяются в группы и производится агрегация.
Затем HAVING — отфильтровываются группы, соответствующие условию.
После этого SELECT — отбираются указанные столбцы.
Потом ORDER BY — производится сортировка результирующей таблицы.
И в самом конце LIMIT — ограничивается количество выводимых записей.

Разбейте заказы из таблицы orders на 3 группы в зависимости от количества товаров, попавших в заказ:

Малый (от 1 до 3 товаров);
Средний (от 4 до 6 товаров);
Большой (7 и более товаров).
Посчитайте число заказов, попавших в каждую группу. Группы назовите соответственно «Малый», «Средний», «Большой» (без кавычек).

Выведите наименования групп и число товаров в них. Колонку с наименованием групп назовите order_size, а колонку с числом заказов — orders_count.

Отсортируйте полученную таблицу по колонке с числом заказов по возрастанию.

```sql
SELECT 
    CASE
        WHEN array_length(product_ids, 1) >= 1 AND array_length(product_ids, 1) <= 3 THEN 'Малый'
        WHEN array_length(product_ids, 1) >= 4 AND array_length(product_ids, 1) <= 6 THEN 'Средний'

    COUNT(order_id) AS orders_count

    orders
GROUP BY 
    order_size
ORDER BY 
    orders_count ASC;
```

По данным из таблицы orders рассчитайте средний размер заказа по выходным и будням.

Группу с выходными днями (суббота и воскресенье) назовите «weekend», а группу с будними днями (с понедельника по пятницу) — «weekdays» (без кавычек).

В результат включите две колонки: колонку с группами назовите week_part, а колонку со средним размером заказа — avg_order_size. 

Средний размер заказа округлите до двух знаков после запятой.

Результат отсортируйте по колонке со средним размером заказа — по возрастанию.

Поля в результирующей таблице: week_part, avg_order_size

<br>
<br>
<br>
<br>

<span style="color: darkgreen;" >
<center>Ответ: </center>
<br>

```sql
SELECT case when to_char(creation_time, 'Dy') in ('Sat', 'Sun') then 'weekend'
            else 'weekdays' end as week_part,
       round(avg(array_length(product_ids, 1)), 2) as avg_order_size
FROM   orders
GROUP BY week_part
ORDER BY avg_order_size
```

Для каждого пользователя в таблице user_actions посчитайте общее количество оформленных заказов и долю отменённых заказов.

Новые колонки назовите соответственно orders_count и cancel_rate. Колонку с долей отменённых заказов округлите до двух знаков после запятой.

В результат включите только тех пользователей, которые оформили больше трёх заказов и у которых показатель cancel_rate составляет не менее 0.5.

Результат отсортируйте по возрастанию id пользователя.

<span style="color: darkgreen;" >
<center>Ответ: </center>
<br>

```sql
SELECT user_id,
       count(distinct order_id) as orders_count,
       round((count(distinct order_id) filter (WHERE action = 'cancel_order')::numeric / count(distinct order_id)),
             2) as cancel_rate
FROM   user_actions
GROUP BY user_id having count(distinct order_id) > 3 and (count(distinct order_id) filter (
WHERE  action = 'cancel_order')::numeric / count(distinct order_id)) >= 0.5
ORDER BY user_id asc;
```
